```python
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Carregar o dataset
df = pd.read_csv('/content/recurso__6867029_xml.csv')
```

```python
# Selecionar as colunas relevantes
features = ['ASSENTOS_SALA', 'ASSENTOS_CADEIRANTES', 'ASSENTOS_MOBILIDADE_REDUZIDA', 'ASSENTOS_OBESIDADE']
df_selected = df[features].dropna()
```

```python
# Normalizar os dados
df_normalized = (df_selected - df_selected.mean()) / df_selected.std()
```

```python
# Aplicar PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df_normalized)
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
```

```python
# Plotar o gráfico de dispersão
plt.figure(figsize=(8,6))
plt.scatter(principal_df['PC1'], principal_df['PC2'], c='blue', alpha=0.5)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - Análise de Componentes Principais')
plt.grid()
plt.show()
```

```python
# Informações sobre PCA
explained_variance = pca.explained_variance_ratio_
cov_matrix = pca.get_covariance()
eigenvalues = pca.explained_variance_
eigenvectors = pca.components_
```

```python
print("Variância explicada pelos componentes principais:", explained_variance)
print("Matriz de Covariância:", cov_matrix)
print("Autovalores:", eigenvalues)
print("Autovetores:", eigenvectors)
```
