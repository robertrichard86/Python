#Importando

```python
import pandas as pd
import nltk
import unicodedata
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import os
```

#Processando

```python
# Função para limpar texto
def limpar_texto(texto):
    if not isinstance(texto, str) or not texto.strip():
        return None

    # Remove acentos
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    # Minúsculas
    texto = texto.lower()
    # Remove pontuação
    texto = texto.translate(str.maketrans('', '', string.punctuation))
    # Tokeniza
    tokens = word_tokenize(texto)
    # Remove stopwords
    stop_words = set(stopwords.words('portuguese'))
    tokens = [t for t in tokens if t not in stop_words]
    # Rejunta
    return ' '.join(tokens) if tokens else None

# Função principal
def processar_arquivo_txt(caminho_txt):
    if not os.path.isfile(caminho_txt) or not caminho_txt.endswith('.txt'):
        print("❌ Caminho inválido ou arquivo não é .txt")
        return

    # Lê todas as linhas
    with open(caminho_txt, 'r', encoding='utf-8') as f:
        linhas = f.readlines()

    # Remove quebras de linha, espaços extras
    linhas = [linha.strip() for linha in linhas if linha.strip()]

    df = pd.DataFrame(linhas, columns=["texto_original"])
    df.drop_duplicates(inplace=True)
    df.dropna(subset=["texto_original"], inplace=True)

    # Aplica limpeza
    df['texto_limpo'] = df['texto_original'].apply(limpar_texto)
    df.dropna(subset=["texto_limpo"], inplace=True)
    df = df[df['texto_limpo'].str.strip().astype(bool)]

    # Salva apenas a coluna 'texto_limpo' em um arquivo .txt
    output_filename = "dataset_limpo.txt"
    with open(output_filename, 'w', encoding='utf-8') as f:
        for item in df['texto_limpo']:
            f.write("%s\n" % item)

    print(f"\n✅ Dataset limpo salvo como '{output_filename}' com {len(df)} registros.")


# === Execução principal ===
if __name__ == '__main__':
    print("Insira o caminho do arquivo .txt com os textos (um por linha):")
    caminho = input("Ex: meus_textos.txt\n> ").strip()
    processar_arquivo_txt(caminho)
```
