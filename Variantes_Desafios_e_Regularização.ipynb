#Otimização e Regularização em Deep Learning

Este notebook explora as mecânicas que permitem às redes neurais convergirem de forma eficiente e as técnicas para garantir que o modelo generalize bem para dados novos.

## 1. Variantes do Gradiente Descendente
O objetivo de qualquer otimizador é encontrar os pesos $w$ que minimizam a função de perda $J(w)$.

### a) Gradiente Descendente Padrão (Vanilla GD / Batch GD)
Calcula o gradiente da função de perda em relação a **todo o conjunto de dados** antes de realizar uma única atualização nos pesos.
* **Conceito:** É estável e converge para o mínimo global em funções convexas, mas é extremamente lento e exige muita memória para datasets grandes.
* **Equação:** $w = w - \eta \cdot \nabla_w J(w)$

### b) Gradiente Descendente Estocástico (SGD)
Atualiza os pesos após processar **cada exemplo** individualmente.
* **Conceito:** Muito mais rápido e permite aprendizado online. O "ruído" nas atualizações ajuda o modelo a saltar de mínimos locais, mas a convergência é oscilatória e instável.

### c) Momentum
Adiciona uma fração da atualização do passo anterior ao passo atual.
* **Conceito:** Funciona como a inércia de uma bola rolando um morro; ganha velocidade em direções constantes e amortece oscilações laterais bruscas.
* **Equação:** $v_t = \gamma v_{t-1} + \eta \nabla_w J(w) \implies w = w - v_t$

### d) Adam (Adaptive Moment Estimation)
Combina os conceitos de **Momentum** e **RMSProp**, calculando taxas de aprendizado adaptativas para cada parâmetro.
* **Conceito:** É o padrão atual da indústria por ser robusto. Ele mantém médias móveis tanto do gradiente (primeiro momento) quanto do quadrado do gradiente (segundo momento).

## 2. Desafios Comuns em Treinamento

O treinamento de redes profundas enfrenta barreiras físicas e matemáticas que podem impedir a convergência.

### a) Overfitting (Sobreajuste)
Ocorre quando o modelo se torna tão complexo que "decora" o ruído e os detalhes específicos dos dados de treino em vez de aprender o padrão geral.
* **Sintoma:** A perda (loss) de treino continua caindo, mas a perda de validação começa a subir.

### b) Vanishing Gradients (Gradientes Desvanecentes)
Em redes muito profundas, ao realizar o *backpropagation*, o gradiente é multiplicado sucessivamente por valores pequenos (derivadas das funções de ativação como Sigmoid).
* **Conceito:** O gradiente torna-se infinitesimal nas camadas iniciais, fazendo com que os primeiros neurônios parem de aprender.

### c) Exploding Gradients (Gradientes Explosivos)
O oposto do desvanecimento: os gradientes acumulam valores imensos durante a retropropagação, resultando em atualizações de pesos gigantescas.
* **Sintoma:** O modelo torna-se instável e os pesos atingem valores `NaN` (Not a Number).

## 3. Técnicas de Regularização e Estabilização

Para mitigar os desafios acima, aplicamos restrições à rede durante o treinamento.

### a) Dropout
Zera aleatoriamente uma porcentagem (ex: 50%) das saídas dos neurônios durante cada iteração de treino.
* **Objetivo:** Evitar a co-adaptação, forçando a rede a aprender representações redundantes e mais robustas. É uma técnica poderosa contra o Overfitting.

### b) Regularização L2 (Weight Decay)
Adiciona uma penalidade à função de perda baseada na magnitude dos pesos: $L = J(w) + \lambda \sum w^2$.
* **Objetivo:** Penaliza pesos muito grandes, forçando a rede a ser "mais simples" e suave, o que melhora a generalização.

### c) Batch Normalization
Normaliza as entradas de cada camada para que tenham média zero e variância unitária dentro de cada mini-batch.
* **Objetivo:** Estabiliza o aprendizado, permite o uso de taxas de aprendizado mais altas e atua como um regulador leve, combatendo o problema de gradientes desvanecentes.
